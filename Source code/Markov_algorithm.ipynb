```python
import markov_clustering as mc
import networkx as nx
import matplotlib.pyplot as plt
import pickle
```

```python
# Load network data
G = nx.Graph()
data = open("facebookNet.txt","r")
lines=data.readlines()
for line in lines:
    line_split=line.split(" ")

    G.add_edge(int(line_split[0]),int(line_split[1]))
```

```python
# Build adjacency matrix
edge_mat = nx.to_numpy_array(G)
```

```python
#the loop below aims to increase the size of the top 10 nodes in TOPSIS
list1=[107,1684,1912,3437,0,1085,698,567,58,428]
size=[]
for i in G:
    if i in list1:
        size.append(1000)
    else:
        size.append(35)
```

```python
# Run MCL algorithm
result = mc.run_mcl(edge_mat)
clusters = mc.get_clusters(result)
```

```python
plt.figure(figsize=(15,8))
mc.draw_graph(edge_mat, clusters, node_size=size, with_labels=False,edgecolors='black')
```

```python
dict1={}#this dictionary will be fill by community number and number of nodes its have
for i in range(1,len(clusters)+1):
  dict1[i]=len(clusters[i-1])
```

```python
fig = plt.figure(figsize = (15,8))
#draw the plot Nombre of nodes for each community
plt.bar(list(dict1.keys()), list(dict1.values()), color ='green',width = 0.5)

plt.ylabel("Number of community")
plt.ylabel("Nombre of nodes")
plt.title("Nombre of nodes for each community ")
plt.show()
```
