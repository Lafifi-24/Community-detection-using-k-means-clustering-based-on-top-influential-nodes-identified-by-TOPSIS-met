```python
from cdlib import algorithms, viz
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G =nx.read_edgelist("facebookNet.txt")
#implement the louvain algorithm
coms = algorithms.louvain(G)
#draw graph
viz.plot_network_clusters(G,coms,node_size=35,figsize=(15,8),plot_labels=False)
```

```python
len(coms.communities)#number of community by louvain algorithm
```

```python
dict1={}#this dictionary will be fill by community number and number of nodes its have
for i in range(1,len(coms.communities)+1):
  dict1[i]=len(coms.communities[i-1])
```

```python
import matplotlib.pyplot as plt


fig = plt.figure(figsize = (15,8 ))
#draw the plot Nombre of nodes for each community
plt.bar(list(dict1.keys()), list(dict1.values()), color ='green',width = 0.5)

plt.ylabel("Number of community")
plt.ylabel("Nombre of nodes")
plt.title("Nombre of nodes for each community ")
plt.show()
```
